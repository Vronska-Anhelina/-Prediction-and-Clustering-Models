```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn.cluster import DBSCAN

browsing_time = [
    15.5, 16.2, 14.8, 15.3, 16.7, 27.0, 16.5, 15.8, 14.9, 16.1, 25.2, 24.8, 26.0,
    25.5, 16.2, 24.9, 25.7, 25.4, 24.6, 35.8, 40.3, 41.2, 39.5, 40.6, 39.8, 40.2,
    41.0, 40.7, 39.9, 20.1, 50.0, 80.0, 90.0
]

purchase_amount = [
    50.1, 48.5, 51.0, 49.3, 52.7, 90.9, 53.1, 49.5, 50.2, 48.8, 150.3, 152.0,
    151.2, 249.8, 150.6, 151.0, 150.5, 152.3, 149.9, 151.5, 100.1, 202.3, 199.5,
    201.6, 200.8, 199.7, 201.2, 202.0, 200.5, 301.1, 10.0, 500.0, 600.0
]

data=pd.DataFrame({"browsing_time":browsing_time,
                   "purchase_amount": purchase_amount})
print(data)
```

```python
dbscan = DBSCAN(eps=30,min_samples = 5)
labels = dbscan.fit_predict(data)
plt.figure(figsize=(8,6))
unique_labels = set(labels)
for label in unique_labels:
    mask = labels == label
    if label == -1:
        plt.scatter(data[mask].iloc[:, 0], data[mask].iloc[:, 1],
                    color='grey', label='Noise', alpha=0.5)
    else:
        plt.scatter(data[mask].iloc[:, 0], data[mask].iloc[:, 1],
                    label=f"Cluster {label+1}", alpha=0.8)

plt.title("DBSCAN Clustering")
plt.xlabel("Browsing time")
plt.ylabel("Purchase amount")
plt.legend()
plt.show()

```

DBSCAN identified 3 clusters and several noise points. The first cluster is shoppers with short viewing times (15–18 min) and small purchases. The second is medium time and average spending. The third is longer viewing times and higher purchases (~200). The algorithm identified noise points with abnormally high purchase amounts or an atypical time-to-spend ratio.
