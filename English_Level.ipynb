```python
import numpy as np
import pandas as pd

categories = ["Elementary", "Pre-Intermediate", "Intermediate", "Upper-Intermediate", "Advanced"]

np.random.seed(2024)

employee_id = ["e00" + str(x) for x in range(1, 21)]
english_level = [np.random.choice(categories) for i in range(20)]

data = pd.DataFrame({
    "employee_id": employee_id,
    "english_level": english_level
})
print(data)
```

```python
level_mapping={"Elementary":0,
"Pre-Intermediate":1,
"Intermediate":2,
"Upper-Intermediate":3,
"Advanced":4}
data["level_encoded"]=data["english_level"].map(level_mapping)
print(data)
```
